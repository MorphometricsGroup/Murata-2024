```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
mask_cood_0_list = []
for i in range(8):
    mask_cood_0_list.append(np.loadtxt('real_data/GmJMC025_02_mask/0_{}.csv'.format(i), delimiter=',', dtype=np.int16))
```

```python
zero_img = np.zeros((3456,5184),dtype=np.int8)
for j in range(len(mask_cood_0_list)):
    zero_img[mask_cood_0_list[j][0],mask_cood_0_list[j][1]] += 255
```

```python
%matplotlib inline
fig = plt.figure(figsize = (20, 20))
fig.patch.set_alpha(0.)
# 3DAxesを追加
ax = fig.add_subplot(111)
plt.imshow(zero_img)
```
