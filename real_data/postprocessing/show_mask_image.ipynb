```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
```

```python
def get_mask_img(ind, img_num, output_folder="."):

    file_list = glob.glob('../masks/{}_mask/{}_*.png'.format(ind, str(img_num)))
    images = []
    for file_name in file_list:
        img = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)
        images.append(img)

    img = np.ones((images[0].shape[0], images[0].shape[1], 3), dtype=np.uint8)*255

    cmap = plt.cm.tab20
    for i, image in enumerate(images):
        img[image == 255] = np.array(cmap(i)[:3][::-1])*255
        
    output_path = output_folder+"/"+str(img_num)+"_mask.jpg"

    cv2.imwrite(output_path, img)
    
def make_nums(ind):
    view_mat_folder = "../view_mats/"+ind+"_view_mat/"
    view_mat_file_list = glob.glob(view_mat_folder+"[0-9]*csv")
    num_list = []
    for view_mat in view_mat_file_list:
        num_list.append(int(view_mat.split("\\")[-1].split(".")[0]))
    return num_list
```

```python
inds = ["exp2021_005_GmJMC112_01_MVI_01",
       "exp2021_005_GmJMC112_01_MVI_02",
       "exp2021_005_GmJMC112_01_MVI_04"]
```

```python
for ind in inds:
    img_nums = make_nums(ind)
    os.makedirs("mask_images/"+ind, exist_ok=True)
    for i in img_nums:
        get_mask_img(ind, i, "mask_images/"+ind)
```
