```python
import glob
import pathlib
import numpy as np
import pickle
import os
```

```python
p = pathlib.Path('temp/')
l = list(p.glob("*.TDlines"))
```

```python
l = [i.stem for i in l]
```

```python
sl = [i.split("_") for i in l]
```

```python
ol = []
for j in range(265):
    templ = []
    templ2 = []
    for i in sl:
        if int(i[0]) == j:
            templ.append(int(i[1]))
            templ2.append(i)

    il = np.argsort(templ)
    ol += list(np.array(templ2)[il])
```

```python
A = [((int(i[0]), int(i[1])), i[2]) for i in ol]
```

```python
l = []
for i in A:
    if i[1] == "F":
        l.append(i)
```

```python
size_list = []
for tag in l:
    size_list.append(os.path.getsize("temp/{0}_{1}_{2}.TDlines".format(tag[0][0],tag[0][1],tag[1])))
```

```python
np.array(size_list)<sum(size_list)/len(size_list)
```

```python
A = np.array(l)[np.where(np.array(size_list)<sum(size_list)/len(size_list))]
```

```python
A.tolist()
```

```python
with open('tag.pickle', 'wb') as f:
    pickle.dump(A.tolist(), f)
```
